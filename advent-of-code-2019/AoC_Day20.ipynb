```python
import numpy as np

data = raw_d.split('\n')
n = len(data[0])
empty = ' '*n
data.insert(0,empty)
data.append(empty)
data = [list(' '+line+' ') for line in data]
data = np.matrix(data)
ABC = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'   

LOC = {}
PORTALS = {}
for j in range(data.shape[0]):
    for i in range(data.shape[1]):        
        if data[j,i] == '.':
            LOC[(j,i)] = set()
            if data[j-1,i] == '.':
                LOC[(j,i)].add((j-1,i))
            if data[j+1,i] == '.':
                LOC[(j,i)].add((j+1,i))
            if data[j,i-1] == '.':
                LOC[(j,i)].add((j,i-1))
            if data[j,i+1] == '.':
                LOC[(j,i)].add((j,i+1))
        if data[j,i] in ABC:
            if data[j-1,i] in ABC and data[j+1,i] == '.':
                if data[j-1,i]+data[j,i] in PORTALS:
                    PORTALS[data[j-1,i]+data[j,i]].add((j+1,i))
                else:
                    PORTALS[data[j-1,i]+data[j,i]] = set([(j+1,i)])
            if data[j+1,i] in ABC and data[j-1,i] == '.':
                if data[j,i]+data[j+1,i] in PORTALS:
                    PORTALS[data[j,i]+data[j+1,i]].add((j-1,i))
                else:
                    PORTALS[data[j,i]+data[j+1,i]] = set([(j-1,i)]) 
            if data[j,i-1] in ABC and data[j,i+1] == '.':
                if data[j,i-1]+data[j,i] in PORTALS:
                    PORTALS[data[j,i-1]+data[j,i]].add((j,i+1))
                else:
                    PORTALS[data[j,i-1]+data[j,i]] = set([(j,i+1)])
            if data[j,i+1] in ABC and data[j,i-1] == '.':
                if data[j,i]+data[j,i+1] in PORTALS:
                    PORTALS[data[j,i]+data[j,i+1]].add((j,i-1))
                else:
                    PORTALS[data[j,i]+data[j,i+1]] = set([(j,i-1)])
end = list(PORTALS.pop('ZZ'))[0]
init = list(PORTALS.pop('AA'))[0]
for portal in PORTALS:
    item = PORTALS[portal]
    a, b = item
    LOC[a].add(b)
    LOC[b].add(a)
```

```python
def bfs_paths(graph, start, goal):
    queue = [(start, [start])]
    while queue:
        (vertex, path) = queue.pop(0)
        for next in graph[vertex] - set(path):
            if next == goal:
                yield path + [next]
            else:
                queue.append((next, path + [next]))
                
def shortest_path(graph, start, goal):
    try:
        return next(bfs_paths(graph, start, goal))
    except StopIteration:
        return Nonekeys 
    
len(shortest_path(LOC, init, end))-1
```

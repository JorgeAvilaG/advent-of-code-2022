```python
data = '''F10
N3
F7
R90
F11'''
data = [(line[0], int(line[1:])) for line in data.split('\n')]
data
```

```python
d = ['N', 'E', 'S', 'W']
direct = 'E'
posd = 1
pN = 0
pE = 0

for code,n in data:
    if code == 'F':
        code = direct
    if code == 'N':
        pN += n
    elif code == 'S':
        pN -= n
    elif code == 'E':
        pE += n
    elif code == 'W':
        pE -= n
    elif code == 'R':
        
        posd = int((posd+n/90)%4)
        direct = d[posd]
    elif code == 'L':
        posd = int((posd-n/90)%4)
        direct = d[posd]

print(pN,pE,abs(pN)+abs(pE))
        
```

```python
# part 2

d = ['N', 'E', 'S', 'W']
direct = 'E'
posd = 1

pN_B = 0
pE_B = 0
pN_W = 1
pE_W = 10


for code,n in data:
    if code == 'F':
        pN_B += n*pN_W
        pE_B += n*pE_W
    if code == 'N':
        pN_W += n
    elif code == 'S':
        pN_W -= n
    elif code == 'E':
        pE_W += n
    elif code == 'W':
        pE_W -= n
    elif code == 'R':
        temp = n/90
        if temp == 1:
            pE_W, pN_W = pN_W, -pE_W
        elif temp == 2:
            pE_W, pN_W = -pE_W, -pN_W
        elif temp == 3:
            pE_W, pN_W = -pN_W, pE_W
            
    elif code == 'L':
        temp = n/90
        if temp == 3:
            pE_W, pN_W = pN_W, -pE_W
        elif temp == 2:
            pE_W, pN_W = -pE_W, -pN_W
        elif temp == 1:
            pE_W, pN_W = -pN_W, pE_W
    
print(pN_B,pE_B, abs(pN_B)+abs(pE_B))
        

```

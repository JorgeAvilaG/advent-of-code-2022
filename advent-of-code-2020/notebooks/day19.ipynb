```python
import numpy as np
f = open("../data/day19.txt", "r")
data = [line.strip().split(':') for line in list(f)]
D ={}
for line in data:
    if line[0].isnumeric():
        D[int(line[0])] = [i.split(' ') for i in line[1].strip().split(' | ')]
        
D[35] = ['a']
D[72] = ['b']
```

```python
def apply_rule(rule,solutions):
    caracters = solutions

    for r in D[rule][0]:
        if r == '35':
            caracters = [x + 'a' for x in caracters]   
        elif r == '72':
            caracters = [x + 'b' for x in caracters]
        else:
            caracters = [x + y for x in caracters for y in apply_rule(int(r),[''])]

    
        
    if len(D[rule]) > 1:
        caracters2 = solutions
        for r in D[rule][1]:
            if r == '35':
                caracters2 = [x + 'a' for x in caracters2]   
            elif r == '72':
                caracters2 = [x + 'b' for x in caracters2]
            else:
                caracters2 = [x + y for x in caracters2 for y in apply_rule(int(r),[''])]
                
                
                
        return caracters + caracters2
    else:
        return caracters
        

#solutions = apply_rule(0,[''])   
```

```python
#sol part1
solutions = apply_rule(0,[''])   
sol = 0
solutions = set(solutions)
f = open("../data/day19.txt", "r")

for line in list(f):
    if line.strip() in solutions:
        sol +=1
print(sol)
```

```python
#sol part2
solutions = apply_rule(31,[''])  
s_31 = set(solutions)
solutions = apply_rule(42,[''])  
s_42 = set(solutions)
```

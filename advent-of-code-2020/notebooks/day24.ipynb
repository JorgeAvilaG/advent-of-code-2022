```python
f = open("../data/day24.txt", "r")
data = [line.strip() for line in list(f)]
data
```

```python
def find_place(line):
    x = 0
    y = 0
    temp = ''
    for pos in line:
        temp += pos
        if temp == 'e':
            x -= 1
            temp = ''
        elif temp == 'w':
            x+=1
            temp = ''
        elif temp == 'nw':
            y+=1
            temp = ''
        elif temp == 'se':
            y-=1
            temp = ''
        elif temp == 'ne':
            x-=1
            y+=1
            temp = ''
        elif temp == 'sw':
            x+=1
            y-=1
            temp = ''
    return x,y

black = set()
for line in data:
    x,y = find_place(line)
    if (x,y) not in black:
        black.add((x,y))
    else:
        black.remove((x,y))
black    
```

```python
## part 2


def nears(pos):
    x,y = pos
    return [(x,y+1),(x,y-1),(x+1,y),(x-1,y),(x-1,y+1),(x+1,y-1)]      
    


```

```python
for x in range(100):
    cand_black = set()
    for pos in black:
        temp = nears(pos)
        for t in temp:
            cand_black.add(t)

    #to_black
    to_black = set()
    for pos in cand_black:
        near = nears(pos)
        total = 0
        for n in near:
            if n in black:
                total+=1
        if total == 2:
            to_black.add(pos)

    #to_white
    to_white = set()
    for pos in black:
        near = nears(pos)
        total = 0
        for n in near:
            if n in black:
                total+=1
        if total > 2 or total == 0:
            to_white.add(pos)

    for pos in to_white:
        black.remove(pos)
    for pos in to_black:
        black.add(pos)
    print(len(black))
```

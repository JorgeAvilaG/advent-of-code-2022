```python
import numpy as np
f = open("../data/day7.txt", "r")

D = {}
for line in f:
    k = line.strip().split(' bags ')[0]
    D[k] = []
    v = line.strip().split(' contain')[1]
    v = v.split(',')
    for item in v:
        temp = item.strip().split(' ')
        if temp[0] != 'no':
            n = int(temp[0])
            bag = temp[1] +' '+ temp[2]
            D[k].append((bag,n))
D
```

```python
# part 1

sol = 0
for bag in D.keys():
    to_do = [(bag,0)]
    while True:
        current = to_do.pop()
        temp = D[current[0]]
        if 'shiny gold' == current[0]:
            sol += 1
            break
        to_do.extend(temp)
        if len(to_do) == 0:
            break   

sol-1
```

```python
#part 2

current = ('shiny gold', 1)

def part_2(two):
    current, n = two[0] ,two[1]
    if D[current] == []:
        return n
    else:
        a = 0
        for item in D[current]:
            a += part_2(item)
        return n+n*a

part_2(current)-1
```

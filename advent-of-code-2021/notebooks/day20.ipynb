```python
import numpy as np
f = open("../data/data_day20.txt", "r")
data = list(f)
data = [line.strip() for line in data]
algorithm = data[0]
data = [list(line) for line in data[2:]]
data = np.array(data)
    
```

```python
algorithm
```

```python
def get_enhancement(pos,data):
    x,y = pos
    sol = ''
    for i in [-1,0,1]:
        for j in [-1,0,1]:
            if data[y+i][x+j] == '#':
                sol+='1'
            else:
                sol+='0'
    return algorithm[int(sol,2)]

def increase_size(data,paint):
    temp = []
    temp.append(list(paint*(len(data)+2)))
    for line in data:
        temp.append(np.concatenate((np.array([paint]),line,np.array([paint]))))
    temp.append(list(paint*(len(data)+2)))
    temp = np.array(temp)
    return temp

def print_data(data):
    sol = 0
    for i in range(len(data)):
        for j in range(len(data)):
            if data[i][j] == '#': sol+=1 
            print(data[i][j], end='')
        print('')
    print(sol)    
    
    
```

```python
# part 1 n = 2
# part 2 n = 50
n = 50

for counter in range(n):
    if counter%2 == 0:
        paint = '.'
    else:
        paint = '#'
    
    data = increase_size(data,paint).copy()
    data = increase_size(data,paint).copy()
    data = increase_size(data,paint).copy()
    #print_data(data)
    new_data = data.copy()
    new_data = []
    for y in range(1,len(data)-1):
        line = []
        for x in range(1,len(data)-1):
            temp = get_enhancement((x,y),data)
            line.append(temp)
            #new_data[y][x] = temp
        new_data.append(line)
    data = new_data.copy()

print_data(data)
    
```

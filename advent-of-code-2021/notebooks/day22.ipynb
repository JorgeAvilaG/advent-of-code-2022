```python
f = open("../data/data_day22.txt", "r")
data = list(f)
data = [line.strip().split(' ') for line in data]
new_data = []
for line in data:
    x,y,z = line[1].split(',')
    x = x[2:].split('..')
    y = y[2:].split('..')
    z = z[2:].split('..')
    x = [int(n) for n in x]
    y = [int(n) for n in y]
    z = [int(n) for n in z]
    new_data.append([line[0],x,y,z])

new_data[0:11]
    
```

```python
# part 1
on = set()
for line in new_data[:]:
    
    if line[1][0] >= 50 or line[1][0] <= -50:
        break
    
    if line[0] == 'on':
        to_on = True
    else:
        to_on = False
        
    for x in range(line[1][0],line[1][1]+1):
        for y in range(line[2][0],line[2][1]+1):
            for z in range(line[3][0],line[3][1]+1):
                if to_on:
                    on.add((x,y,z))
                elif (x,y,z) in on:
                    on.remove((x,y,z))
                
        
print(len(on))  
```

```python
# part 2
def check_if_(new_cube, set_of_cubes):
    more_cubes = []
    for cube in set_of_cubes:
        cube_x = cube[0]
        new_cube_x = new_cube[0]
        cube_y = cube[1]
        new_cube_y = new_cube[1]
        cube_z = cube[2]
        new_cube_z = new_cube[2]
        new_x = max(cube_x[0], new_cube_x[0])
        new_x2 = min(cube_x[1], new_cube_x[1])
        new_y = max(cube_y[0], new_cube_y[0])
        new_y2 = min(cube_y[1], new_cube_y[1])
        new_z = max(cube_z[0], new_cube_z[0])
        new_z2 = min(cube_z[1], new_cube_z[1])
        if new_x > new_x2 or new_y > new_y2 or new_z > new_z2:
            continue
        more_cubes.append(((new_x,new_x2),(new_y,new_y2),(new_z,new_z2)))

    return more_cubes

```

```python
on_dimensions = set()
off_dimensions = set()

for line in new_data[:]:

    new_cube = (tuple(line[1]),tuple(line[2]),tuple(line[3]))

    more_off = check_if_(new_cube, on_dimensions)
    more_on = check_if_(new_cube, off_dimensions)
    
    cheaking = set()
    for candidate in more_off:
        if candidate in more_on:
            more_on.remove(candidate)
        else:
            cheaking.add(candidate)
    more_off = cheaking

    for candidate in more_off:
        if not candidate in on_dimensions:
            off_dimensions.add(candidate)
        else:
            on_dimensions.remove(candidate)
    for candidate in more_on:
        if not candidate in off_dimensions:
            on_dimensions.add(candidate)
        else:
            off_dimensions.remove(candidate)

    if line[0] == 'on':
        if not new_cube in off_dimensions:
            on_dimensions.add(new_cube)
        else:
            off_dimensions.remove(new_cube)

    print(len(on_dimensions),len(off_dimensions))
               
        
```

```python
total = 0
for x,y,z in on_dimensions:
    total += (abs(x[1]-x[0])+1) * (abs(y[1]-y[0])+1) * (abs(z[1]-z[0])+1)
for x,y,z in off_dimensions:
    total -= (abs(x[1]-x[0])+1) * (abs(y[1]-y[0])+1) * (abs(z[1]-z[0])+1)
print(total)
```

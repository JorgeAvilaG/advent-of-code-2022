```python
f = open("../data/data_day23.txt", "r")
data = list(f)
data = [list(line.strip()) for line in data]
data
```

```python
maze = set()
hallways = set()
positions = [0,0,0,0,0,0,0,0]
for y in range(len(data)):
    for x in range(len(data[y])):
        if data[y][x] == ".":
            maze.add((x, y))
            hallways.add((x,y))
        if data[y][x] in "ABCD":
            maze.add((x, y))
            if data[y][x] == "A":
                if positions[0] == 0:
                    positions[0] = (x, y)
                else:
                    positions[1] = (x, y)
            if data[y][x] == "B":
                if positions[2] == 0:
                    positions[2] = (x, y)
                else:
                    positions[3] = (x, y)
            if data[y][x] == "C":
                if positions[4] == 0:
                    positions[4] = (x, y)
                else:
                    positions[5] = (x, y)
            if data[y][x] == "D":
                if positions[6] == 0:
                    positions[6] = (x, y)
                else:
                    positions[7] = (x, y)

maze
positions
hallways
good = set()
for h in hallways:
    if not h[0] in [3,5,7,9]:
        good.add(h)
hallways = good
hallways
```

```python
def check_end(p):
    if p[0] in [(3,2),(3,3)] and p[1] in [(3,2),(3,3)]:
        if p[2] in [(5,2),(5,3)] and p[3] in [(5,2),(5,3)]:
            if p[4] in [(7,2),(7,3)] and p[5] in [(7,2),(7,3)]:
                if p[6] in [(9,2),(9,3)] and p[7] in [(9,2),(9,3)]:
                    return True
    return False
```

```python
from collections import deque
def bfs(start, end):
    queue = deque()
    queue.append((start, 0))
    visited = set()
    visited.add(start)
    while queue:
        pos, steps = queue.popleft()
        if pos == end:
            return steps
        for d in [(0,1),(0,-1),(1,0),(-1,0)]:
            new_pos = (pos[0]+d[0], pos[1]+d[1])
            if new_pos in maze and new_pos not in visited:
                if new_pos not in p:
                    queue.append((new_pos, steps+1))
                    visited.add(new_pos)
    return -1

def bfs2(start, end):
    queue = deque()
    queue.append((start, 0))
    visited = set()
    visited.add(start)
    while queue:
        pos, steps = queue.popleft()
        if pos == end:
            return steps
        for d in [(0,1),(0,-1),(1,0),(-1,0)]:
            new_pos = (pos[0]+d[0], pos[1]+d[1])
            if new_pos in maze and new_pos not in visited:
                queue.append((new_pos, steps+1))
                visited.add(new_pos)
    return 0

def proyected_energy(p):
    energy = 0
    for i in range(8):
        if i == 0:
            if p[i] == (3,3):
                continue
            if p[i] == (3,2) and p[1] == (3,3):
                continue
            energy += bfs2(p[i], (3,3))*1
        elif i == 1:
            if p[i] == (3,3):
                continue
            if p[i] == (3,2) and p[0] == (3,3):
                continue
            energy += bfs2(p[i], (3,3))*1
        elif i == 2:
            if p[i] == (5,3):
                continue
            if p[i] == (5,2) and p[3] == (5,3):
                continue
            energy += bfs2(p[i], (5,3))*10
        elif i == 3:
            if p[i] == (5,3):
                continue
            if p[i] == (5,2) and p[2] == (5,3):
                continue
            energy += bfs2(p[i], (5,3))*10
        elif i == 4:
            if p[i] == (7,3):
                continue
            if p[i] == (7,2) and p[5] == (7,3):
                continue
            energy += bfs2(p[i], (7,3))*100
        elif i == 5:
            if p[i] == (7,3):
                continue
            if p[i] == (7,2) and p[4] == (7,3):
                continue
            energy += bfs2(p[i], (7,3))*100
        elif i == 6:
            if p[i] == (9,3):
                continue
            if p[i] == (9,2) and p[7] == (9,3):
                continue
            energy += bfs2(p[i], (9,3))*1000
        elif i == 7:
            if p[i] == (9,3):
                continue
            if p[i] == (9,2) and p[6] == (9,3):
                continue
            energy += bfs2(p[i], (9,3))*1000
    return energy
```

```python
f = open("../data/data_day24.txt", "r")
data = list(f)
data = [line.strip() for line in data]
data = [line.split(' ') for line in data]



def inp(a, b):
    return b
def add(a, b):
    return a+b
def mult(a, b):
    return a*b
def div(a, b):
    return a//b
def mod(a,b):
    return (a%b)
def eql(a,b):
    if a == b: return 1
    else: return 0
def gen(number):
    for x in number:
        yield int(x)

F = {
    'inp' : inp,
    'add' : add,
    'mul' : mult,
    'div' : div,
    'mod': mod,
    'eql': eql,    
}    
V = {
    'x':0,
    'y':1,
    'z':2,
    'w':3,
}
MONAD = []
for line in data:
    part1 = F[line[0]]
    part2 = line[1]
    if len(line) != 3:
        MONAD.append([part1,part2,'input'])
        continue
    if line[2] in 'xyzw':
        part3 = line[2]
    else:
        part3 = int(line[2])
    MONAD.append([part1,part2,part3])

MONAD[:19]
```

```python
f = open("../data/data_day4.txt", "r")
data = list(f)
data = [x.strip() for x in data]
numbers = data[0]
numbers = numbers.split(',')
numbers = [int(n) for n in numbers]
print(numbers[0:10])
print(data[1:50])

i = -1
Boards = {}
Scores = {}
for line in data[1:]:
    if line == '':
        i+=1
        Boards[i] = []
        Scores[i] = []
    else:
        Boards[i].extend([int(n) for n in line.split()])
        Scores[i].extend([0,0,0,0,0])
        
```

```python
def mark_scores(n):
    for i in Boards.keys():
        for j, number in enumerate(Boards[i]):
            if number == n:
                Scores[i][j] = 1
        
def check_winning():
    for i in Scores.keys():
        score = Scores[i]
        for n in range(0,len(score),5):
            if sum(score[n:n+5]) == 5:
                
                return True,i
        for n in range(0,5):
            sol = 0
            nn = n
            while nn < 25:
                sol += score[nn]
                nn+=5
            if sol == 5:
                return True,i
    return False,i 


def cal_solution(board):
    sol = 0
    for j, number in enumerate(Scores[board]):
        if number == 0:
            sol+=Boards[board][j]
    return sol
                
        

```

```python
# part 1
for n in numbers:
    mark_scores(n)
    if check_winning()[0]:
        board = check_winning()[1]
        winner = n
        break
sol = cal_solution(board)
print(sol,winner)
winner*sol
    
```

```python
# part 2
winning = 0
for n in numbers:
    mark_scores(n)
    while True:
        win,i= check_winning()
        print(win,winning,len(Scores.keys()))
        if win:
            winning += 1
        else:
            break
        if winning < 100: 
            del Boards[i]
            del Scores[i]
        if winning > 100:
            break
        if winning == 100:
            board = i
            winner = n
            
            sol = cal_solution(board)
            print(sol,winner)
            print(winner*sol)
            break
```

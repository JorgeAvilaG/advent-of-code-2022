```python
f = open(".\data\day16.txt", "r")
data = [x.strip() for x in list(f)]
f.close()

import re
D = {}
for line in data:
    pairs = re.findall(r'([A-Z]{2})', line)
    D[pairs[0]] = (int(re.findall(r'(\d+)',line)[0]), set(pairs[1:]))
D
```

```python
def distance(a,b):
    to_check = [(a,0)]
    visited = {}
    while to_check:
        current, distance = to_check.pop(0)
        if current in visited:
            if distance >=visited[current]:
                continue
        visited[current] = distance
        if current == b:
            return distance
        for x in D[current][1]:
            to_check.append((x, distance+1))
```

```python
Distances = {}
for room in D.keys():
    for other in D.keys():
        if room == other:
            continue
        Distances[(room, other)] = distance(room, other)
Distances
    
```

```python
#part 1
interesting_rooms = []
for room in D.keys():
    if D[room][0] > 0:
        interesting_rooms.append(room)

to_check = []
to_check.append(('AA',30,0,interesting_rooms.copy()))
solutions = 0
counter = 0
while len(to_check) > 0:
    #counter += 1
    #if counter%5000 == 0:
        #print(len(to_check), solutions)
    
    room, time, punt, interesting_rooms = to_check.pop()
    if time == 0:
        if punt > solutions:
            solutions = punt
        continue
    if len(interesting_rooms) == 0:
        if punt > solutions:
            solutions = punt
        continue

    for next_room in interesting_rooms:
        dist = Distances[(room, next_room)]#distance(room,next_room)
        if dist+1 <= time:
            new_interesting_rooms = interesting_rooms.copy()
            new_interesting_rooms.remove(next_room)
            new_time = time - dist - 1
            new_punt = punt + D[next_room][0]*(new_time)
            to_check.append((next_room,new_time,new_punt,new_interesting_rooms))

print(solutions)
    
```

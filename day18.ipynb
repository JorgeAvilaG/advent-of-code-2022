```python
f = open(".\data\day18.txt", "r")
data = [x.strip().split(',') for x in list(f)]
f.close()
data = [[int(x) for x in line] for line in data]
data[0:10]
```

```python
#part 1
rocks = [tuple(line) for line in data]


sol = 0
for a,b,c in data:
    counter = 0
    for x in [a+1, a-1]:
        if (x,b,c) in rocks:
            counter += 1
    for x in [b+1, b-1]:
        if (a,x,c) in rocks:
            counter += 1
    for x in [c+1, c-1]:
        if (a,b,x) in rocks:
            counter += 1
    sol+=6-counter
print(sol)
```

```python
#part 2
air_inside = {}

sol_part1 = 0
for a,b,c in data:
    counter = 0
    for x in [a+1, a-1]:
        if (x,b,c) in rocks:
            counter += 1
        elif (x,b,c) not in air_inside:
            air_inside[(x,b,c)] = 1
        else:
            air_inside[(x,b,c)] += 1
    for x in [b+1, b-1]:
        if (a,x,c) in rocks:
            counter += 1
        elif (a,x,c) not in air_inside:
            air_inside[(a,x,c)] = 1
        else: 
            air_inside[(a,x,c)] += 1
    for x in [c+1, c-1]:
        if (a,b,x) in rocks:
            counter += 1
        elif (a,b,x) not in air_inside:
            air_inside[(a,b,x)] = 1
        else:
            air_inside[(a,b,x)] += 1
    sol_part1+=6-counter
print(sol_part1)
```

```python
max_x = max([x[0] for x in data])
max_y = max([x[1] for x in data])
max_z = max([x[2] for x in data])
min_x = min([x[0] for x in data])
min_y = min([x[1] for x in data])
min_z = min([x[2] for x in data])
print(max_x, max_y, max_z)
print(min_x, min_y, min_z)
```

```python
def is_inside(pos):
    visited = set()
    to_check = [pos]
    while to_check:
        x,y,z = to_check.pop()
        if x > 19 or y >19  or z > 19:
            return False
        elif x < 0 or y < 0 or z < 0:
            return False
        visited.add((x,y,z))
        for a,b,c in [(x-1,y,z),(x,y-1,z),(x,y,z-1),(x+1,y,z),(x,y+1,z),(x,y,z+1)]:
            if (a,b,c) in rocks:
                continue
            if (a,b,c) not in visited:
                to_check.append((a,b,c))
    return True

sol_part2 = sol_part1
counter = 0
for air in air_inside.keys():
    if is_inside(air):
        counter += 1
        print(counter, air)
        sol_part2 = sol_part2 - air_inside[air]
print(sol_part2)
```

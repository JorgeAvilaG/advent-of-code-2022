```python
# part 1
def more_objectives(to_check, new_resources, new_robots, new_time, blueprint):
    n=0
    if new_robots[0] > 0:
        if  new_robots[1] == 0 or new_robots[0] < max(blueprint[1], blueprint[2], blueprint[3], blueprint[5])+n:
            to_check.append((tuple(new_resources), tuple(new_robots), new_time, 0))
        if  new_robots[2] == 0 or new_robots[1] < blueprint[4]+n:
            to_check.append((tuple(new_resources), tuple(new_robots), new_time, 1))
    if new_robots[0] > 0 and new_robots[1] > 0:
        if new_robots[3] == 0 or new_robots[2] < blueprint[6]+n:
            to_check.append((tuple(new_resources), tuple(new_robots), new_time, 2))
    if new_robots[0] > 0 and new_robots[2] > 0:
        to_check.append((tuple(new_resources), tuple(new_robots), new_time, 3))
    return to_check
```

```python
import math
def part1(blueprint):
    solution = []
    blueprint = blueprint
    resources = (0,0,0,0)
    robots = (1,0,0,0)
    time = 24
    to_check = []
    to_check.append((resources, robots, time, 0)) # 0 = objective is to produce the robot 0
    to_check.append((resources, robots, time, 1)) # 1 = objective is to produce the robot 1

    while to_check:
        resources, robots, time, objective = to_check.pop()
        #print(resources, robots, time, objective)
        if time == 0:
            solution.append(resources[-1])
            #print(resources, robots, time, objective)
            continue

        if objective == 0:
            resources_needed = blueprint[1] - resources[0]
            new_resources = list(resources)
            new_robots = list(robots)
            new_time = time
            if resources_needed > 0:
                time_needed = math.ceil(resources_needed/robots[0])
                                            
                if time - time_needed <= 0: # not enoght time to process the robot
                    for n in range(len(robots)):
                        new_resources[n] += robots[n]*time
                    new_time = 0
                    solution.append(resources[-1])
                    #print(new_resources, new_robots, new_time, objective)
                    continue
                for n in range(len(robots)):
                    new_resources[n] += robots[n]*time_needed
                new_time = time - time_needed
            
            for n in range(len(robots)):
                new_resources[n] += robots[n]*1
            new_time -= 1
            new_robots[0] += 1
            new_resources[0] -= blueprint[1]

            to_check = more_objectives(to_check, new_resources, new_robots, new_time, blueprint)

        
           

        if objective == 1:
            resources_needed = blueprint[2] - resources[0]
            new_resources = list(resources)
            new_robots = list(robots)
            new_time = time
            if resources_needed > 0:
                time_needed = math.ceil(resources_needed/robots[0])
                                            
                if time - time_needed <= 0: # not enoght time to process the robot
                    for n in range(len(robots)):
                        new_resources[n] += robots[n]*time
                    new_time = 0
                    solution.append(resources[-1])
                    #print(new_resources, new_robots, new_time, objective)
                    continue
                for n in range(len(robots)):
                    new_resources[n] += robots[n]*time_needed
                new_time = time - time_needed
            
            for n in range(len(robots)):
                new_resources[n] += robots[n]*1
            new_time -= 1
            new_robots[1] += 1
            new_resources[0] -= blueprint[2]
    
            to_check = more_objectives(to_check, new_resources, new_robots, new_time, blueprint)


        if objective == 2:
            resources_needed1 = blueprint[3] - resources[0]
            resources_needed2 = blueprint[4] - resources[1]
            new_resources = list(resources)
            new_robots = list(robots)
            new_time = time
            if resources_needed1 > 0 or resources_needed2 > 0:
                time_needed1 = math.ceil(resources_needed1/robots[0])
                time_needed2 = math.ceil(resources_needed2/robots[1])
                time_needed = max(time_needed1, time_needed2)                          
                if time - time_needed <= 0: # not enoght time to process the robot
                    for n in range(len(robots)):
                        new_resources[n] += robots[n]*time
                    new_time = 0
                    solution.append(resources[-1])
                    #print(new_resources, new_robots, new_time, objective)
                    continue
                for n in range(len(robots)):
                    new_resources[n] += robots[n]*time_needed
                new_time = time - time_needed
            
            for n in range(len(robots)):
                new_resources[n] += robots[n]*1
            new_time -= 1
            new_robots[2] += 1
            new_resources[0] -= blueprint[3]
            new_resources[1] -= blueprint[4]

            to_check = more_objectives(to_check, new_resources, new_robots, new_time, blueprint)

        if objective == 3:
            resources_needed1 = blueprint[5] - resources[0]
            resources_needed2 = blueprint[6] - resources[2]
            new_resources = list(resources)
            new_robots = list(robots)
            new_time = time
            if resources_needed1 > 0 or resources_needed2 > 0:
                time_needed1 = math.ceil(resources_needed1/robots[0])
                time_needed2 = math.ceil(resources_needed2/robots[2])
                time_needed = max(time_needed1, time_needed2)                          
                if time - time_needed <= 0: # not enoght time to process the robot
                    for n in range(len(robots)):
                        new_resources[n] += robots[n]*time
                    new_time = 0
                    solution.append(resources[-1])
                    #print(new_resources, new_robots, new_time, objective)
                    continue
                for n in range(len(robots)):
                    new_resources[n] += robots[n]*time_needed
                new_time = time - time_needed
            
            for n in range(len(robots)):
                new_resources[n] += robots[n]*1
            new_time -= 1
            new_robots[3] += 1
            new_resources[0] -= blueprint[5]
            new_resources[2] -= blueprint[6]

            to_check = more_objectives(to_check, new_resources, new_robots, new_time, blueprint)
        
        solution.append(new_resources[-1]+new_robots[-1]*new_time)
    return max(solution)         
         
    
```
